```python
const_cols = [col for col in df.columns if df[col].nunique() == 1]
print("Columnas constantes:", const_cols)
```

```python
null_cols = [col for col in df.columns if df[col].isna().all()]
print("Columnas completamente nulas:", null_cols)
```

```python
# Identificar columnas no binarias
no_binarias = [col for col in df.columns if df[col].nunique() > 2]

# Rellenar valores nulos en columnas no binarias con la media (si son numéricas)
for col in no_binarias:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df[col].fillna(df[col].mean())
    else:
        df[col] = df[col].fillna(df[col].mode()[0])  # Si es categórica, usa la moda

print("Valores nulos restantes por columna:")
print(df.isna().sum())
```

```python
# Rellenar nulos en columnas binarias con la moda
binarias = [col for col in df.columns if df[col].nunique() == 2]
for col in binarias:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Valores nulos restantes por columna:")
print(df.isna().sum())
```
